```python
#importando a base de dados sobre a covid-19
import pandas as pd
covid_19 = pd.read_csv('C:\Cursos\MBA\caso_full.csv')
```

```python
#listando as informações
covid_19.head(100)
```

```python
#limpando os dados
df_total_br= covid_19.query('place_type =="state"')
```

```python
#listando os dados mais organizados
df_total_br.head()
```

```python
#gerando o total de mortes no dia 07-07-2021 no BRASIL
total_br=df_total_br["new_deaths"].sum()
```

```python
#trazendo o resultado - total de mortes no Brasil até 07-07-2021
print(total_br)
```
